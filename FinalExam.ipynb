# แก้ค่า Missing และเปลี่ยน type ของข้อมูล

# user

```python
alldb = alldb[(alldb['user'].astype(str).str.isdigit()) & (alldb['user'].str.len() == 5)]
alldb
```

# site

```python
alldb = alldb[(alldb['site'].astype(str).str.isdigit()) & (alldb['site'].str.len() == 5)]
alldb
```

# PartitionKey

```python
alldb = alldb[(alldb['PartitionKey'].astype(str).str.isdigit()) & (alldb['PartitionKey'].str.len() == 5)]      # เลือกมาเฉพาะ คอลัมน์ site ที่เป็นตัวเลข & มีตัวเลข เท่ากับ 5 ตัว
alldb
```

# date

```python
alldb['date'] = pd.to_datetime(alldb['date'])     # เปลี่ยนชนิดข้อมูล และจัดรูปแบบข้อมูล
alldb
```

# แก้ type ข้อมูล

```python
alldb['user'] = pd.to_numeric(alldb['user'], errors='coerce') #errors=’coerce’, you’ll transform the non-numeric values into NaN.
alldb['site'] = pd.to_numeric(alldb['site'], errors='coerce')
alldb['ov'] = pd.to_numeric(alldb['ov'], errors='coerce')
alldb['PartitionKey'] = pd.to_numeric(alldb['PartitionKey'], errors='coerce')
alldb['age'] = pd.to_numeric(alldb['age'], errors='coerce')
alldb['phone']=alldb['phone'].astype('string')
alldb['RowKey']=alldb['RowKey'].astype('string')
alldb['urine']=alldb['urine'].astype('string')
```

```python
alldb.isnull().any()
```
